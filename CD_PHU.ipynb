Whitby là thành phố của Durham Region, Guelph là thành phố ở vùng Wellington còn Fergus thì là cộng đồng trong Wellington. <br>
Vì dữ liệu City trong PublicHealthUnit trùng với dữ liệu City CasesReport nên ta sẽ lấy dữ liệu trong PublicHealthUnit. <br>
Do đó ta sẽ thay thế Durham Region và Fergus bằng **Whitby và Guelph**. Đồng thời chỉnh sửa PHURegion (ReportingPHU Của Guelph).

```python
df_phugr['PHUCity'] = df_phugr['PHUCity'].replace('Durham Region','Whitby')
df_phugr['PHUCity'] = df_phugr['PHUCity'].replace('Fergus','Guelph')
```
