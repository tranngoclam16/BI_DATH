```python
import numpy as np
import pandas as pd
```

```python
#'Public health unit.xlsx'
phu = pd.read_excel('PublicHealthUnit.xlsx', engine='openpyxl', sheet_name = 'Sheet1')
phu
```

```python
caserp = pd.read_csv('Cases Report.csv')
caserp
```

```python
phu = phu.astype({"Reporting_PHU_Latitude": float, "Reporting_PHU_Longitude": float})
phu.info()
```

```python
lat = pd.DataFrame(caserp, columns=['Reporting PHU', 'PHU Latitude', 'PHU Longitude'])
lat = lat.rename(columns={'Reporting PHU':'Reporting_PHU'})
lat.drop_duplicates( inplace=True)
lat
```

```python
caserp[caserp['Reporting PHU'].isin(phu.Reporting_PHU)]['PHU Latitude'].values
```

```python
result = pd.merge(phu, lat, how="inner", on=["Reporting_PHU"])
result.info()
```

```python
result['Reporting_PHU_Latitude']=result['PHU Latitude']
result['Reporting_PHU_Longitude']=result['PHU Longitude']
```

```python
result.drop(columns='PHU Latitude', inplace=True)
result.drop(columns='PHU Longitude', inplace=True)
```

```python
result['CreatedDate']=pd.datetime.now()
result['UpdatedDate']=pd.datetime.now()
result.columns
```

```python
result.drop_duplicates(inplace=True)
```

```python
phu['Status']=1
phu.to_excel('PublicHealthUnit.xlsx',index=False)
```

```python
result.columns=['PHUID', 'ReportingPHU', 'ReportingPHUAddress',
       'ReportingPHUCity', 'ReportingPHUPostalCode',
       'ReportingPHUWebsite', 'ReportingPHULatitude',
       'ReportingPHULongitude', 'CreatedDate', 'UpdatedDate']
result.to_excel('PublicHealthUnit.xlsx',index=False)
```
